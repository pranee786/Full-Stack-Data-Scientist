```python
import re
from datetime import datetime
```

```python
date_values = df[0].tolist()
```

```python
def formatting_dates(date_value):
    try:
        # Considering the patterns based on data
        pattern1 = "[A-Z]{1}[a-z]{2}\s\d{1,2}[,]*\s\d{2,4}"
        pattern2 = "\d{1,2}[,]*\s[A-Z]{1}[a-z]{2}\s\d{2,4}"
        
        formatted_dates_list = []
        
        for value in date_values:
            match1 = re.findall(pattern1,value)
            splitted_values = []
            # If Pattern1 matches exactly with the value
            if len(match1) != 0:
                # Splitting the value using space
                splitted_values = value.replace(',','').split(' ')
            else:
                # If Pattern1 doesnot match, then check with Pattern2
                match2 = re.findall(pattern2,value)
                if len(match2) != 0:
                    # Splitting the value using space
                    splitted_values = value.replace(',','').split(' ')

            day = ''
            month = ''
            year = ''
            if len(splitted_values) != 0:
                for i in splitted_values:
                    # Checking whether is alpha or digit
                    # If alpha, it is month
                    # If digit, then check whether it is day or year
                    if i.isalpha():
                        month = i
                        
                    elif i.isdigit():
                        # If digit is 1, then it is day
                        if len(i) == 1:
                            day = f'0{i}'
                            
                        elif len(i) == 2:
                            # If digit is 2, and value has prefix 0, then it is day, else it is year
                            if i.startswith('0'):
                                day = i
                            else:
                                year = f'20{i}'
                        elif len(i) == 4:
                            year = i

                string = f'{day}-{month}-{year}'
                new_date = datetime.strptime(string,"%d-%b-%Y")
                formatted_date = datetime.strftime(new_date,"%d/%m/%Y")
                formatted_dates_list.append(formatted_date)
        return formatted_dates_list
        
    except Exception as e:
            print(f'There is an exception {e}')
```

```python
formatted_dates_list = formatting_dates(date_values)

if len(formatted_dates_list) == len(df.index):
    df['Formatted Date'] = formatted_dates_list
```
